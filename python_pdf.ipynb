### export text using Template

```python
from reportlab.platypus import SimpleDocTemplate
from reportlab.platypus import Paragraph
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import PageBreak
from reportlab.lib.units import mm, inch

#
flowables = []
#get default style
sample_style_sheet = getSampleStyleSheet()
# if you want to see all the sample styles, this prints them
sample_style_sheet.list()

#add data into flowables
paragraph_1 = Paragraph("Chapter 1: A title", sample_style_sheet['Heading1'])
paragraph_2 = Paragraph("Some normal body text", sample_style_sheet['BodyText'])
flowables.append(paragraph_1)
flowables.append(paragraph_2)
#add page break and start another paragraph
flowables.append(PageBreak())
paragraph = Paragraph("Chapter 2: A title", sample_style_sheet['Heading1'])
flowables.append(paragraph)

custom_body_style = sample_style_sheet['BodyText']
for f in range(2,40,2):
    custom_body_style.fontSize = f
    paragraph = Paragraph(f"fontsize={f}", custom_body_style)
    flowables.append(paragraph)


#export: path, pagesize, side margin 
pagesize= (11*inch, 8.5*inch) # width, height
my_doc = SimpleDocTemplate('hello.pdf', pagesize=pagesize, 
    topMargin=1*inch,    leftMargin=1*inch,    rightMargin=1*inch,    bottomMargin=1*inch)
my_doc.build(flowables)
```

### export table

coordinate is (column, row). the top left cell (0,0), the bottom right cell (-1,-1)
the top row in a table (0,0),(-1,0)


```python
#fontsize is working
from reportlab.lib import colors
from reportlab.lib.pagesizes import letter, A4, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle

# container for the 'Flowable' objects
elements = []
data= [['00', '01', '02', '03', '04'], ['10', '11', '12', '13', '14'],
        ['20', '21', '22', '23', '24'], ['30', '31', '32', '33', '34']]
t=Table(data)
t.setStyle(TableStyle([
    ('BACKGROUND',(1,1),(-2,-2),colors.green),
('TEXTCOLOR',(0,0),(1,-1),colors.red), ('FONTSIZE', (0,0), (-1,-1), 2)]))
elements.append(t)

#create doc object
doc = SimpleDocTemplate("simple_table.pdf", pagesize=landscape(A4))
# write the document to disk
doc.build(elements)
```

```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import Paragraph, Table, TableStyle
from reportlab.lib.enums import TA_JUSTIFY, TA_LEFT, TA_CENTER
from reportlab.lib import colors

# bodytext  style used for wrapping  data on flowables 
styles = getSampleStyleSheet()
styleN = styles["BodyText"]
#used alignment if required
styleN.alignment = TA_LEFT

styleBH = styles["Normal"]
styleBH.alignment = TA_CENTER


hdescrpcion = Paragraph('''<b>descrpcion</b>''', styleBH)
hpartida = Paragraph('''<b>partida</b>''', styleBH)


descrpcion = Paragraph('long long long long long long long long long long long long line ', styleN)
partida = Paragraph('1', styleN)

data= [[hdescrpcion, hpartida],
       [partida ,descrpcion]]

table = Table(data)

table.setStyle(TableStyle([
                       ('INNERGRID', (0,0), (-1,-1), 0.25, colors.black),
                       ('BOX', (0,0), (-1,-1), 0.25, colors.black),
                       ]))

c = canvas.Canvas("a.pdf", pagesize=A4)
table.wrapOn(c, 50, 50)
table.drawOn(c, 100,600)
c.save()
```
