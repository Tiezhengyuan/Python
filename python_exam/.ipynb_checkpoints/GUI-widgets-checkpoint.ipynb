## button
button = tk.Button(window, command, justify, state)

Button property
command:	the callback being invoked when the button is clicked
justify:	the way in which the inner text is justified: possible (self-describing) values are: LEFT, CENTER, and RIGHT
state:	if you set the property to DISABLED, the button becomes deaf and doesn’t react to clicks, while its title is shown in gray; setting it to NORMAL restores normal button functioning; when the mouse is located above the button, the property changes its value to ACTIVE

Button method
flash():	the button flashes a few times but doesn’t change its state
invoke():	activates the callback assigned to the widget and returns the same value the callback returned; note: this is the only way to invoke your own callback explicitly, as the event manager must be aware of the fact

```python
import tkinter as tk

#bind callback
def click():
    if button['text']=='ok':
        button.config(text='yes')
    else:
        button['text']='ok'

#instantiate button and set property
root = tk.Tk()
button = tk.Button(root, text='ok', command=click)
button.pack()
root.mainloop()
```

```python
import tkinter as tk

#two approaches for changing state
def click():
    if button2['state']=='normal':
        button2.config(state='disabled')
        button3.config(state=tk.DISABLED)
    else:
        button2['state']='normal'
        button3['state']=tk.NORMAL
    
root = tk.Tk()

button = tk.Button(root, text='switch', command=click)
button.pack()
button2 = tk.Button(root, text='start')
button2.pack()
button3 = tk.Button(root, text='start')
button3.pack()

root.mainloop()
```

## checkbutton
use strinVar, and callback function
1. create checkbutton: 
2. set default value
3. bind value with StringVar or IntVar
4. enable and disable: c.config(state='normal')


Checkbutton property
bd:	the checkbutton frame width (default is two pixels)
command:	the callback being invoked when the checkbutton changes its state
justify:	the same as for Button
state:	the same as for Button
variable:	an observable IntVar variable reflecting the widget’s state; defaultly it’s set to 1 when the checkbutton is checked, and to 0 otherwise
offvalue:	the non-default value being assigned to a variable when the checkbutton is not checked
onvalue:	the non-default value being assigned to a variable when the checkbutton is checked


Checkbutton method
deselect()	unchecks the widget
flash()	the same as for Button
invoke()	the same as for Button
select()	checks the widget
toggle()	toggles the widget (changes its state to the opposite one)

```python
#bind StringVar with checked/unchecked

import tkinter as tk

def agg():
    tk.messagebox.showinfo('StringVar', var3.get())


window = tk.Tk()

#tk.Checkbutton
var3=tk.StringVar(window)
#set default value: uncheck
var3.set(False)
#onvalue and offvalue set value of var3
cb=tk.Checkbutton(window, text='yes', variable=var3, 
    onvalue='good', offvalue='bad', command=agg)
cb.pack()

window.mainloop()
```

```python
#select/unselect checkbutton
import tkinter  as tk 
my_w = tk.Tk()
my_w.geometry("300x150")
def my_up():
    if(c_v.get()==1):
        c2.select()   
        c3.select()   
        c4.select()   
    else:
        c2.deselect() 
        c3.deselect() 
        c4.deselect() 

def my_up1():
    c2.toggle()
    c3.toggle()
    c4.toggle()
    
def my_s():
    s=[]
    if c2_v.get()==1:
        s.append('apple')
    if c3_v.get()==1:
        s.append('orange')
    if c4_v.get()==1:
        s.append('banana')
    tk.messagebox.showinfo('selected', ', '.join(s))
    
#use select()/deselect()
c_v = tk.IntVar()
c_v.set(False)
c=tk.Checkbutton(my_w,text='All', variable=c_v, onvalue=1, offvalue=0, 
    command=my_up,font=18)
c.grid(row=0,column=1)

#use toggle()
c1_v = tk.IntVar()
c1_v.set(False)
c1=tk.Checkbutton(my_w,text='all_toggle', variable=c1_v, 
    onvalue=1, offvalue=0, command=my_up1,font=18)
c1.grid(row=0,column=2)


c2_v=tk.IntVar(my_w)
c2=tk.Checkbutton(my_w,text='apple',font=18,variable=c2_v)
c2.grid(row=1,column=1)

c3_v=tk.IntVar(my_w)
c3=tk.Checkbutton(my_w,text='orange',font=18,variable=c3_v)
c3.grid(row=1,column=2)

c4_v=tk.IntVar(my_w)
c4=tk.Checkbutton(my_w,text='banana',font=18,variable=c4_v)
c4.grid(row=1,column=3)

b=tk.Button(my_w, text="ok",command=my_s)
b.grid(row=2, column=2)

my_w.mainloop()
```

```python
#enable and disable checkbutton
#checkbutton.config(state)

import tkinter  as tk 
my_w = tk.Tk()
my_w.geometry("300x150")
def my_upd():
    if(c1_v1.get()=='Yes'):
        c2.config(state='disabled')
    else:
        c2.config(state='normal')    

c1_v1=tk.StringVar(my_w)
c1_v1.set('')
c1=tk.Checkbutton(my_w,text='Enable / Disable',variable=c1_v1,
    onvalue='Yes',offvalue='',command=my_upd,font=18)
c1.grid(row=1,column=1)
c2_v1=tk.StringVar(my_w)
c2_v1.set('')
c2=tk.Checkbutton(my_w,text='Python',variable=c2_v1,
                  onvalue='Python',offvalue='',font=18)
c2.grid(row=1,column=2)
my_w.mainloop()
```

## radiobutton

Radiobutton property

Radiobutton method

## Label

Label property
1. text:	a string which will be shown within the Label; note: newline characters (\n) are interpreted in the usual way
2. textvariable:	the same as for text, but makes use of an observable StringVar variable, so if you change the variable’s alteration, it will be immediately visible on the screen.

Label has not methods

```python
import tkinter as tk


def to_string(x):
    return "Current counter\nvalue is:\n" + str(x)


def plus():
    global counter
    counter += 1
    text.set(to_string(counter))


counter = 0
window = tk.Tk()
button = tk.Button(window, text="Go on!", command=plus)
button.pack()
text = tk.StringVar()
label = tk.Label(window, textvariable=text, height=4)
text.set(to_string(counter))
label.pack()
window.mainloop()
```

```python
import tkinter as tk

def plus():
    var.set(var.get()+1)

def minus():
    t=var.get()-1
    var.set(t if t>=1 else 1)

window=tk.Tk()
label_1=tk.Label(window,text='counts:')
label_1.grid(row=1, column=1)

var=tk.IntVar()
var.set(1)
label_2=tk.Label(window,textvariable=var, width=10)
label_2.grid(row=1, column=2, columnspan=2)

button_1 = tk.Button(window, text="+", command=plus)
button_1.grid(row=1, column=4)

button_2 = tk.Button(window, text="-", command=minus).grid(row=1, column=5)


window.mainloop()
```

## Entry
Entry property
1. command:	although Entry is obviously a clickable widget, it doesn’t allow you to bind a callback through the command property. You can observe and control all occurring changes instead by setting the tracer function for the observable variable which cooperates with Entry (we’ll show you this – be patient!)
2. show:	a string assigned to this property will be displayed instead of the actual characters entered into the input field; e.g., if you set show='*', this will enable the widget to safely edit the user’s password
3. state:	the same as for Button
4. textvariable:	an observable StringVar reflecting the current state of the input field
5. width:	the input field’s width (in characters)


Entry method
get():	returns the current input field’s contents as a string
set(s):	sets the whole input field’s contents with the s string
delete(first, last=None):	deletes a part of the input field’s contents; first and last can be integers with values indexing the string; if the last argument is omitted, a single character is deleted; if last is specified as END, it points to the place after the last field’s character
insert(index, s):	inserts the s string at the field position pointed to by index

```python
#default value using insert()

import tkinter as tk
root = tk.Tk()
root.geometry("200x100")

entry = tk.Entry(root)
entry.insert(0, "Default Text")
entry.pack()

root.mainloop()
```

```python
#default value using variable

import tkinter as tk
root = tk.Tk()
root.geometry("200x100")

var=tk.StringVar()
var.set("Default Text")
entry = tk.Entry(root, textvariable=var)
entry.pack()

root.mainloop()
```

```python
#default value using variable

import tkinter as tk

def on_click(event):
    if var.get()=="Default Text":
        event.widget.delete(0, tk.END)

root = tk.Tk()
root.geometry("200x100")

var=tk.StringVar()
var.set("Default Text")
entry = tk.Entry(root, textvariable=var)
entry.bind("<Button-1>", on_click)
entry.pack()

root.mainloop()
```

## Menu
mnu = Menu(master, option, ...)

Let’s summarize the most important menu traits:

a classic menu is actually a horizontal bar located at the top of the application window;
the bar contains a number of horizontally deployed options, often referred to as items or entries;
these options can have hot-keys (keyboard shortcuts enabling the user to quickly access selected operations without using a mouse; usually, hot-keys are triggered by pressing Alt-hotkey on the keyboard)
selecting a menu’s option (it doesn’t matter whether through a hotkey or by a mouse click) causes one of two effects:
it launches a callback bound to the option;
it unrolls a new menu (actually a submenu)
if you want to have such a menu within your Tkinter application, you have to:
create a top-level menu object;
embed it inside the window;
bind a number of required submenus (this is called a cascade) or connect a single callback.

menu = tk.Meanu(window)
window.config(meanu=menu)

submenu = tk.Meanu(menu, tearoff=1)
menu.add_cascade(label='File', menu=submenu, underline=0)

```python
#underline=0,1 set Alt-F, Alt-B as hotkey, 
#tearoff=1, 
import tkinter as tk
from tkinter import messagebox

def open_file():
    messagebox.showinfo("Open doc", "We'll open a file here...")

def open_a_file(file_name):
    messagebox.showinfo("Open doc", f"Open file {file_name}")
    
def are_you_sure():
    if messagebox.askyesno("", "Are you sure you want to quit the App?"):
        window.destroy()
        
def about_app():
    messagebox.showinfo("App", "The application\nthat does nothing")


window = tk.Tk()

# main menu creation
main_menu = tk.Menu(window)
window.config(menu=main_menu)

# 1st main menu item: an empty (as far) submenu
sub_menu_file = tk.Menu(main_menu, tearoff=0)
main_menu.add_cascade(label="File", menu=sub_menu_file, underline=0)

#
sub_menu_recent=tk.Menu(sub_menu_file, tearoff=1)
infiles=['a.txt','b.txt','c.txt']
for f in infiles:
    sub_menu_recent.add_command(label=f, underline=1, 
        command=lambda: open_a_file(f))


# add items to the submenu
sub_menu_file.add_command(label='open', command=open_file)

sub_menu_file.add_cascade(label='recently', menu=sub_menu_recent)
#add sperator between commands
sub_menu_file.add_separator()
#accelerator naming hot-key
sub_menu_file.add_command(label="Quit", accelerator='Ctrl-Q',
    underline=0, command=are_you_sure)
#make a global binding
window.bind_all("<Control-q>", are_you_sure)

# 2nd main menu item: a simple callback
sub_menu_help = tk.Menu(main_menu)
sub_menu_help = tk.Menu(main_menu)
main_menu.add_command(label="About...", command=about_app, underline=1)

window.mainloop()
```

```python
import tkinter as tk


def on_off():
    global accessible
    if accessible == tk.DISABLED:
        accessible = tk.ACTIVE
    else:
        accessible = tk.DISABLED
    sub_menu.entryconfigure(1, state=accessible)


accessible = tk.DISABLED
window = tk.Tk()
menu = tk.Menu(window)
window.config(menu=menu)
sub_menu = tk.Menu(menu, tearoff=0)
menu.add_cascade(label="Menu", menu=sub_menu)
sub_menu.add_command(label="On/Off", command=on_off)
sub_menu.add_command(label="Switch", state=tk.DISABLED)
window.mainloop()
```

### canvas.create_arc(x0,y0,x1,y1,..., style, start, extent)


```python
import tkinter as tk

window = tk.Tk()
canvas = tk.Canvas(window, width=400, height=400, bg='yellow')
canvas.create_arc(10, 100, 380, 300, outline='red', width=5)
canvas.create_arc(10, 100, 380, 300, outline='blue', width=5,
                  style=tk.CHORD, start=90, fill='white')
canvas.create_arc(10, 100, 380, 300, outline='green', width=5,
                  style=tk.ARC, start=180, extent=180)
button = tk.Button(window, text="Quit", command=window.destroy)
canvas.grid(row=0)
button.grid(row=1)
window.mainloop()
```

### canvas.create_oval(x0,y0,x1,y1,..., style, start, extent)
 Creates a circle or an ellipse at the given coordinates. 
It takes two pairs of coordinates; the top left and bottom right corners of the bounding rectangle for the oval.

```python
import tkinter as tk

window = tk.Tk()

canvas = tk.Canvas(window, width=400, height=400, bg='yellow')
canvas.create_oval(10, 10, 30, 30)
canvas.create_oval(40, 40, 100, 100, fill='red', outline='black', width=2)
canvas.create_oval(100, 40, 160, 100, fill='blue')
canvas.pack()

window.mainloop()
```

```python
import tkinter as tk

def change(event):
    global n, colors
    n+=1
    if n>2:
        n=0
    canvas.create_oval(10, 10, 200, 200, fill=colors[n])
    print(event.__dict__)

window = tk.Tk()

canvas = tk.Canvas(window, width=400, height=400, bg='yellow')
canvas.pack()

n=0
colors=['red', 'yellow','green']
canvas.create_oval(10, 10, 200, 200, fill=colors[n])
canvas.bind("<Button-1>", change)


window.mainloop()
```

### canvas.create_text(x0,y0,x1,y1,..., fill, font, justify, text, width)

```python
import tkinter as tk

window = tk.Tk()
canvas = tk.Canvas(window, width=400, height=400, bg='blue')
canvas.create_text(200, 200, text="Mary\nhad\na\nlittle\nlamb",
                   font=("Arial","40","bold"),
                   justify=tk.CENTER,
                   fill='white')
button = tk.Button(window, text="Quit", command=window.destroy)
canvas.grid(row=0)
button.grid(row=1)
window.mainloop()
```

### canvas.create_image(x,y,..., image)

```python
import tkinter as tk

window = tk.Tk()
canvas = tk.Canvas(window, width=400, height=400, bg='yellow')
image = tk.PhotoImage(file='logo.png')
canvas.create_image(200, 200, image=image)
button = tk.Button(window, text="Quit", command=window.destroy)
canvas.grid(row=0)
button.grid(row=1)
window.mainloop()
```

```python
import tkinter as tk
import PIL

window = tk.Tk()
canvas = tk.Canvas(window, width=400, height=400, bg='red')
jpg = PIL.Image.open('logo.jpg')
image = PIL.ImageTk.PhotoImage(jpg)
canvas.create_image(200, 200, image=image)
button = tk.Button(window, text="Quit", command=window.destroy)
canvas.grid(row=0)
button.grid(row=1)
window.mainloop()
```
