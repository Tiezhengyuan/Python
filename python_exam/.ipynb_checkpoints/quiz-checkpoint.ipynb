Your application should keep track of two parameters:

the number of apples processed, stored as a class variable;
the total weight of the apples processed; stored as a class variable. Assume that each apple's weight is random, and can vary between 0.2 and 0.5 of an imaginary weight unit;

```python
class Apple:
    total_number = 0
    total_weight = 0
    
    def __init__(self, number):
        self.ask_number=number
        
    def trace_package(self, weights):
        if self.total_weight > 300 or self.total_number>self.ask_number:
            pass
        else:
            self.total_number +=1
            self.total_weight += weights[0]
            self.trace_package(weights[1:])

import random
weights = [ random.uniform(.2,.5) for _ in range(2000)]
w=Apple(1000)
r=w.trace_package(weights)
print(r)
print(w.total_number)
print(w.total_weight)
```

Create a class representing a time interval;
the class should implement its own method for addition, subtraction on time interval class objects;
the class should implement its own method for multiplication of time interval class objects by an integer-type value;
the __init__ method should be based on keywords to allow accurate and convenient object initialization, but limit it to hours, minutes, and seconds parameters;
the __str__ method should return an HH:MM:SS string, where HH represents hours, MM represents minutes and SS represents the seconds attributes of the time interval object;
check the argument type, and in case of a mismatch, raise a TypeError exception.

```python
class TimeInterval:
    def __init__(self, hr, min, sec):
        try:
            self.hr = int(hr)
        except:
            print("wrong hour")
        self.min=min
        self.sec=sec
        self.t= self.hr*60*60 + self.min*60+self.sec

    def __add__(self, time):
        acc_sec = time.t + self.t
        ihr, imin, isec = self.sec_to_hmr(acc_sec)
        return f"{ihr}:{imin}:{isec}"
            
    def __sub__(self, time):
        inter_sec = time.t - self.t
        ihr, imin, isec = self.sec_to_hmr(inter_sec)
        return f"{ihr}:{imin}:{isec}"
    
    def sec_to_hmr(self, acc_sec):
        ihr = acc_sec//(60*60)
        imin = (acc_sec%(60*60))//60
        isec = acc_sec - ihr*60*60 - imin*60
        return ihr, imin, isec
        
    def __str__(self):
        t = "{:0>2d}:{:0>2d}:{:0>2d}".format(self.hr,self.min, self.sec)
        return t

t1=TimeInterval(3,4,5)
t2=TimeInterval(23,14,10)
print('Time 1: ', t1)
print('Time 2: ', t2)
print(t1-t2)
print(t1+t2)

#t3=TimeInterval('ab23',14,10)
```

```python
class BankError(Exception):
    @staticmethod
    def balanceException():
        raise Exception("wrong balance")
    @staticmethod
    def accountException():
        raise Exception("can't change account number")

class Bank:
    def __init__(self):
        self._account = None
        self._balance = 0
    
    @property
    def account(self):
        return self._account
    
    @account.setter
    def account(self, account):
        if self._account:
            raise BankError.accountException()
        self._account = account
    
    @property
    def balance(self):
        return self._balance
        
    @balance.setter
    def balance(self, balance):
        #print('begain set balance:', balance)
        if balance<0:
            raise BankError.balanceException()
        self._balance=balance
    
    @balance.deleter
    def balance(self):
        self._balance = 0
        self._account = None
    
    def deposit(self, value):
        self.balance +=value

    def withdraw(self, value):
        self.balance -=value
b=Bank()
b.balance=1000;print(b.__dict__)
#b.balance=-300;print(b.balance)
b.account='001';print(b.account)
#b.account='002';print(b.account)
b.deposit(1000);print(b.balance)
del b.balance;print(b.__dict__)
```

```python
import json

class Vehicle:
    def __init__(self):
        self._car = {
            "registration_number": '',
            "year_of_production": 0,
            "passenger": True,
            "mass": 0
        }
        
    @property
    def reg(self):
        return self._car["registration_number"]
    @reg.setter
    def reg(self, reg):
        self._car["registration_number"] = reg

    @property
    def yr(self):
        return self._car["year_of_production"]
    @yr.setter
    def yr(self, yr):
        self._car["year_of_production"] = yr
        
    @property
    def passenger(self):
        return self._car["passenger"]
    @passenger.setter
    def passenger(self, p):
        if p in ("Y", "y"):
            self._car["passenger"] = True
        else:
            self._car["passenger"] = False
        
    @property
    def mass(self):
        return self._car["mass"]
    @mass.setter
    def mass(self, mass):
        self._car["mass"] = mass
        
    def get_json_str(self):
        self.reg = input("Registration number: ")
        self.yr = input("Year of production: ")
        self.passenger = input("Passenger [y/n]: ")
        self.mass = input("vehicle mass: ")
        print("Resulting JSON string is: ", json.dumps(self._car))
        return json.dumps(self._car)
    
    def decode_json_str(self):
        v_str = input("Enter vehicle JSON string: ")
        print(v_str)
        self._car = json.loads(v_str)
        print(self._car)
        return self._car

if __name__ == "__main__":
    print("What can I do for your?")
    print("1- produce a JSON string describing a vehicle")
    print("2- decode a JSON string into vehicle data")
    task = input('Your choice: ')
    if task == '1':
        Vehicle().get_json_str()
    elif task == "2":
        Vehicle().decode_json_str()
    print("Done")

#{"registration_number":45433,"year_of_production":2009,"passenger":false,"mass":234}
```

It's a small excerpt of the New York Stock Exchange quotes list. Take a look at it and analyze its structure. You need to do this as your task is to write a code which reads the data and presents it in a form similar to this one:

```python
import xml.etree.ElementTree as et
import os




infile = "nyse.xml"
#infile = "test1.txt"
try:
    tree = et.parse(infile)
    root = tree.getroot()
    print(tree)
    print(root.tag, root.attrib)
    print("{:<38s}\t{:<8s}\t{:<8s}\t{:<8s}\t{:<8s}".format(\
        "COMPANY", "LAST", "CHANCE", "MIN", "MAX"))
    print("-"*100)
    for child in root:
        print("{:38s}\t{:<8.3f}\t{:<8.3f}\t{:<8.3f}\t{:<8.3f}".format(\
            child.text, float(child.attrib['last']), \
            float(child.attrib['change']), float(child.attrib['min']), \
            float(child.attrib['max'])))
except FileNotFoundError as e:
    print('no file: ', e) 
except et.ParseError as e:
    print('wrong xml: ', e)
```

```python
import sqlite3

class DB:
    def __init__(self, db_file, table_name):
        self.conn = None
        self.table = table_name
        try:
            self.conn = sqlite3.connect(db_file)
            self.c = self.conn.cursor()
        except sqlite3.OperationalError:
            raise Exception("failed in connecting db")
        
    def create_table(self):
        try:
            self.c.execute('''
                create table tasks (
                    id INTEGER PRIMARY KEY,
                    name TEXT NOT NULL,
                    priority INTEGER NOT NULL
                );
            ''')
        except sqlite3.OperationalError as e:
            #raise Exception("Cant create table")
            print("Cant create table", e)
            
    def close_db(self):
        if self.c:
            self.c.close()
        if self.conn:
            self.conn.commit()
            self.conn.close()

    def add_task(self, data):
        try:
            sql = f"INSERT INTO {self.table} (name, priority) VALUES (?,?)"
            self.c.execute(sql, data)
        except sqlite3.OperationalError as e:
            #raise Exception("Cant create table")
            print("Cant add tasks", e)

    def add_tasks(self, data):
        try:
            sql = f"INSERT INTO {self.table} (name, priority) VALUES (?,?)"
            self.c.executemany(sql, data)
        except sqlite3.OperationalError as e:
            #raise Exception("Cant create table")
            print("Cant add tasks:", e)

    def update_task(self, id, key, value):
        try:
            sql = f"UPDATE {self.table} SET {key}=? WHERE id=?"
            self.c.execute(sql, (value, id))
        except sqlite3.OperationalError as e:
            print("Cant update task:", e)
            
    def find_task(self, task_name):
        sql = f"SELECT * FROM {self.table} WHERE NAME=?"
        items =  self.c.execute(sql, (task_name,)).fetchall()
        print(items)
        return items
        
    def show_tasks(self):
        sql = f"SELECT * FROM {self.table}"
        items = self.c.execute(sql)
        for i in items:
            print(i)
        return items
#
d= DB('todo.db', 'tasks')
#d.create_table()
#d.add_task(('east', 3))
#new_tasks = [('cary', 3),('north', 4), ('south',1)]
#d.add_tasks(new_tasks)
#d.find_task('cary')
d.update_task(5, 'name', 'KK')
d.show_tasks()
d.close_db()
print("done")
```
